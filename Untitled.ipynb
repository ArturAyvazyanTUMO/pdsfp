```python
# Run this cell, but please don't change it.

# These lines import the Numpy and Datascience modules.
import numpy as np
from datascience import *

# These lines do some fancy plotting magic
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import warnings
warnings.simplefilter('ignore', FutureWarning)
warnings.simplefilter('ignore', UserWarning)
from matplotlib import patches
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
```

```python
armenia=Table.read_table("tables/armenia.csv")
armenia.show()
luxemburg=Table.read_table("tables/luxemburg.csv")
jordan=Table.read_table("tables/jordan.csv")
israel=Table.read_table("tables/israel.csv")
georgia=Table.read_table("tables/georgia.csv")
liechtenstein=Table.read_table("tables/liechtenstein.csv")
slovenia=Table.read_table("tables/slovenia.csv")
```

```python
def removeComma(clm):
    return np.char.replace(clm,",","")
```

```python
x=armenia.column(0)
y=removeComma(armenia.column(1))

line1, = plt.plot(x, y, label="Armenia's population")
line1.set_dashes([2, 2, 10, 2])  # 2pt line, 2pt break, 10pt line, 2pt break

x=luxemburg.column(0)
y=removeComma(luxemburg.column(1))

line2 = plt.plot(x, y, label="Luxemburg's population")

x=jordan.column(0)
y=removeComma(jordan.column(1))

line4 = plt.plot(x, y,color='#ff0894', label="Jordan's population")

x=israel.column(0)
y=removeComma(israel.column(1))

line5 = plt.plot(x, y, label="Israel's population")

x=georgia.column(0)
y=removeComma(georgia.column(1))

line6 = plt.plot(x, y, label="Georgia's population")

x=slovenia.column(0)
y=removeComma(slovenia.column(1))

line10 = plt.plot(x, y,color="aqua", label="Slovenia's population")

plt.suptitle('Population of countries which are similar to Armenia', fontsize=16)
plt.rcParams['figure.figsize']=[15,10]
plt.legend(loc='lower',title="Population Comparing")
plt.show()
```
